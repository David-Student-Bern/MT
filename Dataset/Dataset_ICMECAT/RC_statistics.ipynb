```python
import pandas as pd
import numpy as np
import astropy
import sunpy
from sunpy.time import parse_time
import matplotlib.pyplot as plt
import seaborn as sns
#make figures bigger
plt.rcParams["figure.figsize"] = (10,5)
plt.rcParams["figure.figsize"] = (10,5)
```

```python
rc = pd.read_csv('RC_icmecat_dates.csv')
```

# Histograms

```python
sns.set_style("darkgrid")

# time between ICME and MO in hours
ICME_start_time = parse_time(rc.ICME_Start)
ICME_end_time = parse_time(rc.ICME_End)

# time between ICME start times
time_between_ICME = ICME_start_time[1:] - ICME_start_time[:-1]

plt.figure(figsize=(12,5))
sns.histplot(time_between_ICME.to('day').value, label='Time between ICME and next ICME',color='mediumseagreen',kde=True)
plt.xlabel('Time difference (days)')
# sns.histplot(time_between_ICME.to('hour').value, label='Time between ICME and next ICME',color='mediumseagreen',kde=True)
# plt.xlabel('Time difference (hours)')
plt.legend(loc=1)
plt.show()

#icme duration
duration = ICME_end_time - ICME_start_time

plt.figure(figsize=(12,5))
# sns.histplot(duration.to('day').value, label='ICME duration',color='mediumseagreen',kde=True)
# plt.xlabel('Duration (days)')
sns.histplot(duration.to('hour').value, label='ICME duration',color='mediumseagreen',kde=True)
plt.xlabel('Duration (hours)')
plt.legend(loc=1)
plt.show()
```
